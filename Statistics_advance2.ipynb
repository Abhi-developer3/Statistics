1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method that allows you to test assumptions (hypotheses) about a population parameter (like the mean or proportion) using sample data.
2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Null Hypothesis (H‚ÇÄ): Assumes no effect or no difference.
- Alternative Hypothesis (H‚ÇÅ): Assumes there is an effect or a difference.
3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (Œ±) in hypothesis testing is the threshold for deciding whether to reject the null hypothesis, commonly set at 0.05. It represents the probability of making a Type I error ‚Äî rejecting a true null hypothesis. A lower Œ± means stricter criteria for claiming a significant result. It is important because it controls the risk of false positives in conclusions.
4. What does a P-value represent in hypothesis testing?
- The probability of observing the test results under the assumption that the null hypothesis is true.
5. How do you interpret the P-value in hypothesis testing?
- P < Œ±: Reject H‚ÇÄ (statistically significant).
- P ‚â• Œ±: Do not reject H‚ÇÄ.
6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type I Error (False Positive): Rejecting H‚ÇÄ when it's true.
- Type II Error (False Negative): Not rejecting H‚ÇÄ when it's false.
7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- In hypothesis testing, the difference between a one-tailed and a two-tailed test lies in the direction of the hypothesis being tested. A one-tailed test is used when you want to determine whether a parameter is either greater than or less than a certain value, focusing on one direction only. For example, you might test whether a new drug increases recovery speed compared to the standard. In contrast, a two-tailed test checks for a difference in either direction, without specifying whether the parameter is higher or lower than the hypothesized value. This is used when any significant change is of interest, regardless of direction. The two-tailed test splits the significance level (Œ±) between both ends of the distribution, while the one-tailed test places the entire Œ± on one side.
8.  What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means. It is used when the population variance is known and the sample size is large (typically n ‚â• 30). The test statistic follows a standard normal distribution (Z-distribution). It is commonly used for testing means and proportions.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is a probability distribution that is similar to the normal distribution but has heavier tails, meaning it is more spread out. It is used when the sample size is small (typically less than 30) and the population standard deviation is unknown. Unlike the normal distribution, the T-distribution accounts for extra uncertainty due to estimating the standard deviation from the sample. You should use the T-distribution instead of the normal distribution when working with small samples and unknown population variance to get more accurate confidence intervals and hypothesis tests.
11. What is the difference between a Z-test and a T-test?
- The main difference between a Z-test and a T-test lies in the information you have about the population and the sample size:
- A Z-test is used when the population standard deviation is known and the sample size is large (usually ùëõ ‚â• 30 n‚â•30). It relies on the normal distribution (Z-distribution).
- A T-test is used when the population standard deviation is unknown and the sample size is small (usually ùëõ < 30 n < 30). It uses the T-distribution ,which adjusts for extra uncertainty from estimating the standard deviation.
12. What is the T-test, and how is it used in hypothesis testing?
- The T-test is a statistical test used to determine whether there is a significant difference between the means of two groups or between a sample mean and a population mean when the population standard deviation is unknown. It is especially useful for small sample sizes (typically less than 30). In hypothesis testing, the T-test helps decide whether to reject the null hypothesis by comparing the calculated T-statistic‚Äîbased on sample data‚Äîto a critical value from the T-distribution, considering the chosen significance level. If the T-statistic falls into the rejection region, it indicates that the observed difference is unlikely due to chance, supporting the alternative hypothesis.
13. What is the relationship between Z-test and T-test in hypothesis testing?
- The Z-test and T-test are closely related statistical tests used in hypothesis testing to compare means, but they differ mainly in how they handle population variability and sample size. The Z-test is used when the population standard deviation is known and the sample size is large, relying on the normal distribution. The T-test is used when the population standard deviation is unknown and the sample size is small, using the T-distribution which adjusts for the added uncertainty. As the sample size grows larger, the T-distribution approaches the normal distribution, so for large samples, the T-test and Z-test give very similar results. Essentially, the T-test can be seen as a generalization of the Z-test for situations with less information about population variability.
14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values, calculated from sample data, that is likely to contain the true population parameter (like the mean) with a certain level of confidence‚Äîcommonly 95%. It provides an estimated range within which we expect the true value to lie, rather than giving a single point estimate. In interpreting statistical results, confidence intervals help show the precision and reliability of the estimate: a narrower interval means more precision, while a wider interval indicates more uncertainty. If a confidence interval for a difference between groups does not include zero, it suggests a statistically significant effect.
15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the amount added and subtracted from a sample estimate to create a confidence interval. It reflects the maximum expected difference between the sample statistic and the true population parameter due to sampling variability. A larger margin of error results in a wider confidence interval, indicating less precision, while a smaller margin of error produces a narrower interval, showing more precise estimates. Essentially, the margin of error controls the range of uncertainty around the estimate within the confidence interval.
16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem is used in statistics to update the probability of a hypothesis based on new evidence or data. It combines prior knowledge (the prior probability) with the likelihood of observed data to calculate a revised probability (the posterior probability). This theorem is significant because it provides a formal way to incorporate new information and improve decision-making under uncertainty, making it fundamental in areas like medical diagnosis, machine learning, and risk assessment. In short, Bayes‚Äô Theorem helps us update beliefs rationally as we gather more data.
17. What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is a probability distribution that describes the sum of the squares of independent standard normal random variables. It is a skewed distribution that depends on degrees of freedom, which usually relate to the number of variables or categories involved. The Chi-square distribution is commonly used in statistics for tests involving categorical data, such as the Chi-square test for independence (to check if two categorical variables are related) and the goodness-of-fit test (to see how well observed data fit an expected distribution). It‚Äôs especially useful when analyzing frequencies or counts in categories.
18. What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness of fit test is a statistical test used to determine whether an observed frequency distribution matches an expected distribution. It helps to check if the data fit a specific theoretical model or hypothesis. To apply the test, you:

- Define the null hypothesis (H‚ÇÄ) stating that the observed data follow the expected distribution.
- Calculate the expected frequencies for each category based on the hypothesized distribution.
- Compute the Chi-square statistic by summing the squared differences between observed and expected frequencies, divided by the expected frequencies.
- Compare the calculated Chi-square value to the critical value from the Chi-square distribution table with appropriate degrees of freedom.
- If the statistic exceeds the critical value, reject H‚ÇÄ, indicating the observed data do not fit the expected distribution well.
19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is a probability distribution that arises when comparing the variances of two independent samples. It is positively skewed and depends on two sets of degrees of freedom‚Äîone for the numerator and one for the denominator. In hypothesis testing, the F-distribution is primarily used in ANOVA (Analysis of Variance) to test whether the means of three or more groups are equal by comparing their variances. It is also used in tests comparing two variances, such as the F-test. The F-distribution helps determine if observed differences between group variances are statistically significant or likely due to random chance.
20. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others. Instead of comparing groups pairwise, ANOVA assesses overall differences by analyzing the variance within and between groups.

Key assumptions of ANOVA are:
- Independence: Observations in each group must be independent of each other.
- Normality: The data in each group should be approximately normally distributed.
- Homogeneity of variances: The variances across groups should be roughly equal (also called homoscedasticity).
21. What are the different types of ANOVA tests?
- One-Way ANOVA: Tests for differences among the means of three or more independent groups based on a single factor or independent variable.
- Two-Way ANOVA: Examines the effect of two independent factors on a dependent variable, including any interaction between the factors.
- Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions or over time.
- MANOVA (Multivariate ANOVA): Extends ANOVA to test for differences in multiple dependent variables simultaneously across groups.
22. What is the F-test, and how does it relate to hypothesis testing?
- The F-test is a statistical test that compares the variances of two or more groups to determine if they are significantly different. It calculates an F-statistic by dividing the variance between group means by the variance within the groups. In hypothesis testing, the F-test is commonly used in ANOVA to check if there are any significant differences among group means. A large F-statistic suggests that the group means are not all equal, leading to rejection of the null hypothesis that all group means are the same.

```python
# 9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std

    plt.scatter(range(len(z_scores)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

sample_data = [10, 12, 23, 23, 16, 23, 21, 16]
zs = calculate_and_plot_zscores(sample_data)
print("Z-scores:", zs)
```

```python
# 14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    return (data_point - mean) / std

data = np.random.normal(50, 10, 1000)
data_point = 65
z = calculate_z_score(data_point, data)
print(f"Z-score of data point {data_point}: {z:.2f}")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Distribution')
plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
plt.title('Z-score on Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 15.Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm
sample = np.array([52, 55, 53, 58, 54, 57, 59, 56, 60, 54])
population_mean = 50
alpha = 0.05
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
# 16.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

data = np.array([52, 55, 53, 58, 54, 57, 59, 56, 60, 54])
confidence = 0.95

n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)

margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
ci_lower = mean - margin
ci_upper = mean + margin

print(f"Sample Mean: {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm

sample = np.array([52, 55, 53, 58, 54, 57, 59, 56, 60, 54])
pop_mean = 50
pop_std = 5
n = len(sample)
alpha = 0.05

sample_mean = np.mean(sample)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

```python
# 2.Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
import numpy as np
from scipy.stats import ttest_1samp

np.random.seed(42)
data = np.random.normal(loc=52, scale=5, size=30)
pop_mean = 50

t_stat, p_value = ttest_1samp(data, pop_mean)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject null hypothesis: No significant difference from population mean.")
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

sample = np.array([52, 55, 53, 58, 54, 57, 59, 56, 60, 54])
pop_mean = 50
pop_std = 5
n = len(sample)
alpha = 0.05

sample_mean = np.mean(sample)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

sample = np.array([52, 55, 53, 58, 54, 57, 59, 56, 60, 54])
pop_mean = 50
pop_std = 5
n = len(sample)
alpha = 0.05

sample_mean = np.mean(sample)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

z_critical = norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')
plt.title('Two-tailed Z-test Decision Regions')
plt.xlabel('Z value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0=0, mu1=1.5, sigma=1, alpha=0.05):
    z_critical = norm.ppf(1 - alpha)
    beta = norm.cdf(z_critical - (mu1 - mu0)/sigma)

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    h0 = norm.pdf(x, mu0, sigma)
    h1 = norm.pdf(x, mu1, sigma)

    plt.plot(x, h0, label='H0 Distribution (mean=mu0)')
    plt.plot(x, h1, label='H1 Distribution (mean=mu1)')

    plt.fill_between(x, 0, h0, where=(x > mu0 + z_critical*sigma), color='red', alpha=0.3, label='Type I Error (Œ±)')
    plt.fill_between(x, 0, h1, where=(x <= mu0 + z_critical*sigma), color='blue', alpha=0.3, label='Type II Error (Œ≤)')

    plt.axvline(mu0 + z_critical*sigma, color='black', linestyle='--', label='Critical Value')
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type I error (Œ±): {alpha}")
    print(f"Type II error (Œ≤): {beta:.4f}")

plot_type1_type2_errors()
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy.stats import ttest_ind

group1 = np.array([20, 22, 19, 24, 20, 21, 23])
group2 = np.array([30, 28, 27, 35, 29, 31, 27])

t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: The groups have significantly different means.")
else:
    print("Fail to reject null hypothesis: No significant difference between group means.")
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

before = np.array([85, 88, 90, 92, 87, 89, 91])
after = np.array([88, 90, 91, 89, 90, 93, 94])

t_stat, p_value = ttest_rel(before, after)

plt.figure(figsize=(6,4))
plt.plot(range(len(before)), before, 'o-', label='Before')
plt.plot(range(len(after)), after, 's-', label='After')
plt.xticks(range(len(before)), [f'Sub {i+1}' for i in range(len(before))])
plt.ylabel('Score')
plt.title('Paired Sample Comparison')
plt.legend()
plt.grid(True)
plt.show()

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: Significant difference between before and after.")
else:
    print("Fail to reject null hypothesis: No significant difference between before and after.")
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy.stats import ttest_1samp, norm

np.random.seed(0)
sample = np.random.normal(loc=52, scale=5, size=30)
pop_mean = 50
pop_std = 5
n = len(sample)
alpha = 0.05

sample_mean = np.mean(sample)

z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_stat)))

t_stat, t_p_value = ttest_1samp(sample, pop_mean)

print(f"Z-test: Z-stat = {z_stat:.3f}, P-value = {z_p_value:.4f}")
print(f"T-test: T-stat = {t_stat:.3f}, P-value = {t_p_value:.4f}")

if z_p_value < alpha:
    print("Z-test: Reject null hypothesis")
else:
    print("Z-test: Fail to reject null hypothesis")

if t_p_value < alpha:
    print("T-test: Reject null hypothesis")
else:
    print("T-test: Fail to reject null hypothesis")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significanc
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin
    return lower, upper

data = [17, 15, 14, 16, 13, 14, 15, 17, 16, 14]
ci_lower, ci_upper = confidence_interval(data)

print(f"Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    return t_crit * std_err

data = [16, 15, 14, 16, 13, 14, 15, 17, 16, 14]
me = margin_of_error(data)

print(f"Margin of Error: {me:.3f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, likelihood_B_given_A, prior_notA, likelihood_B_given_notA):
    numerator = prior_A * likelihood_B_given_A
    denominator = numerator + prior_notA * likelihood_B_given_notA
    posterior_A_given_B = numerator / denominator
    return posterior_A_given_B

prior_A = 0.01
prior_notA = 1 - prior_A
likelihood_B_given_A = 0.99
likelihood_B_given_notA = 0.05
posterior = bayes_theorem(prior_A, likelihood_B_given_A, prior_notA, likelihood_B_given_notA)
print(f"Posterior probability of disease given positive test: {posterior:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in PythoN
import numpy as np
from scipy.stats import chi2_contingency

data = np.array([[25, 15, 20],
                 [30, 16, 35]])

chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-square statistic: {chi2:.3f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print(expected)

if p < 0.05:
    print("Reject null hypothesis: Variables are dependent.")
else:
    print("Fail to reject null hypothesis: Variables are independent.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

def expected_frequencies(observed):
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()
    expected = row_sums @ col_sums / total
    return expected

observed = np.array([[30, 15, 20],
                     [20, 40, 30]])

expected = expected_frequencies(observed)
print(expected)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

observed = np.array([50, 30, 20])
expected = np.array([40, 40, 20])

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject null hypothesis: Observed data fits the expected distribution.")
```
